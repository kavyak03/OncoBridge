# Survival Modeling Example

This notebook demonstrates an educational survival workflow using the merged cohort CSV. If your dataset does not have survival columns, it can add **synthetic demo outcomes** for learning purposes.

```python
import pandas as pd

from ehr_fhir_genomics_toolkit.survival import add_demo_survival_outcomes, fit_cox_example

DATA_PATH = "../demo_sclc_dataset.csv"  # update as needed
df = pd.read_csv(DATA_PATH)
df.head()
```

```python
if {"survival_time_months", "event_observed"} - set(df.columns):
    df = add_demo_survival_outcomes(df, seed=7)
df[["survival_time_months", "event_observed"]].head()
```

```python
cph = fit_cox_example(df)
cph.summary
```
